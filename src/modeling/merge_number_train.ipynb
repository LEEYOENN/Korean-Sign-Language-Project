## 데이터 불러오기

```python
df = pd.read_csv("../../data/merged_num_data.csv")
df
```

```python
df['label'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df.iloc[:, 0] = le.fit_transform(df.iloc[:, 0])
print(le.classes_)
```

## 데이터 분할하기

```python
y = df.iloc[:, 0]
x = df.iloc[:, 1:]

x = df.iloc[:, 64:64+63]
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True, stratify=y)
```

## 모델 학습

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np

model = XGBClassifier(n_estimators=200, max_depth=3, learning_rate=0.1)
#model = RandomForestClassifier(n_estimators=300, max_depth=5, random_state=42)
#model = LogisticRegression(solver='liblinear', max_iter=100)
model.fit(x_train, y_train)

score = model.score(x_test, y_test)

print(f"Accuracy: {np.round(score*100, 2)}%")
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```
