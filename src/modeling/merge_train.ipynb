## 수집한 데이터 합치기

```python
import pandas as pd
merged_data = pd.DataFrame()
for i in range(5, 16):
    data = pd.read_csv(f"../../data/sign_data/sign_data_{i}.csv", header=None)
    merged_data = pd.concat([merged_data, data], ignore_index=True)

merged_data
```

```python
merged_data.to_csv("../../data/merged_data.csv", index=False)
```

## 데이터 불러오기

```python
df = pd.read_csv("../../data/merged_data.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df.iloc[:, 0] = le.fit_transform(df.iloc[:, 0])
print(le.classes_)
```

## 데이터 분할하기

```python
y = df.iloc[:, 0]
x = df.drop(df.columns[0], axis=1)

x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True, stratify=y)
```

## 모델 학습

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# model = XGBClassifier(n_estimators=300, max_depth=5, learning_rate=0.1)
model = RandomForestClassifier(n_estimators=300, max_depth=5, random_state=42)
model.fit(x_train, y_train)

score = model.score(x_test, y_test)

print(f"Accuracy: {np.round(score*100, 2)}%")
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```
